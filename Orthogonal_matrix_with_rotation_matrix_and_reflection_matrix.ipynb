```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as lines
from math import cos, sin, pi, tan
```

```python
def rotate_vectors_by(angle_deg, vecs: list, xlim=[0, 3], ylim=[0, 3]):
    """ 
        Plot vectors which be rotated by the given angle

        Parameters:
            angle_deg: the rotation angle
            vecs: the vectors in array form
            xlim: x boundary
            ylim: y boundary 
    """
    # Change the degree angle unit to radian angle unit
    ang = pi/180*angle_deg;
    # rotation matrix
    rot = np.array([[cos(ang), -sin(ang)], [sin(ang), cos(ang)]])
    # rotated vector = rotation matrix * vector
    vecs_rot = (rot@vecs.T).T

    plt.figure(1)
    ax = plt.gca()

    # plot the vectors that will be rotated in black
    ax.quiver(*np.zeros_like(vecs.T), *vecs.T, 
              angles='xy', scale_units='xy', scale=1,
              color='black')
    # plot the vectors are rotated in blue
    ax.quiver(*np.zeros_like(vecs_rot.T), *vecs_rot.T, 
              angles='xy', scale_units='xy', scale=1,
              color='blue')
    
    # assign vector notation for each vectors
    for v in np.concatenate((vec, vecs_rot)):
        ax.annotate('[{:.2f}, {:.2f}]'.format(v[0], v[1]), (v[0],v[1]), fontsize=12)

    # set frame
    ax.set_xlim(xlim)
    ax.set_ylim(ylim);
    ax.grid()   
```

```python
rotate_vectors_by(30, np.array([[1, 0.5], [2, 0.5]]))
```

```python
def reflect_vectors_acoss_line(angle_deg, vecs: list, xlim=[0, 3], ylim=[0, 3]):
    """ Plot vectors which be reflected by the line 
        that is angled to horizontal axis with the given angle

        Parameters: 
            angle_deg: the angle that reflection line is angled to the horizontal axis
            vecs: the vectors in array form
            xlim: x boundary
            ylim: y boundary         
        """
    # Change the degree angle unit to radian angle unit
    ang = pi/180*angle_deg;
    # reflection matrix
    ref = np.array([[cos(2*ang), sin(2*ang)], [sin(2*ang), -cos(2*ang)]])
    # reflected vector = reflection matrix * vector
    vecs_ref = (ref@vecs.T).T

    plt.figure(1)
    ax = plt.gca()

    # plot the reflection line
    length = np.max(vecs)
    ax.add_artist(lines.Line2D([-length, length], [-length*tan(ang), length*tan(ang)], linestyle='--', color='grey'))
    # plot the vectors that will be reflected by reflection line in black
    
    ax.quiver(*np.zeros_like(vecs.T), *vecs.T, 
              angles='xy', scale_units='xy', scale=1,
              color='black')
    # plot the vectors are reflected by reflection line in blue
    ax.quiver(*np.zeros_like(vecs_ref.T), *vecs_ref.T, 
              angles='xy', scale_units='xy', scale=1,
              color='blue')
    
    # assign vector notation for each vectors
    for v in np.concatenate((vec, vecs_ref)):
        ax.annotate('[{:.2f}, {:.2f}]'.format(v[0], v[1]), (v[0],v[1]), fontsize=12)

    # set frame
    ax.set_xlim(xlim)
    ax.set_ylim(ylim);
    ax.grid()   
```

```python
reflect_acoss_line(45, np.array([[1, 0.5], [2, 0.5]]))
```
